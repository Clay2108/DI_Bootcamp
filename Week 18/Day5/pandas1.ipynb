```python
df=pd.read_csv('all_seasons.csv', encoding='utf-8')  # Load a new dataset
df
```

```python
df.columns  # Display the columns of the DataFrame
```

```python
df2= df[['player_name', 'age','college', 'draft_number', 'team_abbreviation', 'pts']]
df2.tail(50) # Display the last 50 rows of the selected columns  
```

```python
# i want draft number in range 1 to 3 from df2
draft_number_numeric = pd.to_numeric(df2['draft_number'], errors='coerce')
draft_range = df2[(draft_number_numeric >= 1) & (draft_number_numeric <= 3)]
draft_range.tail(30)  # Display the first 10 rows of the filtered DataFrame
```

```python
draft_range.describe()  # Get a statistical summary of the filtered DataFrame
```

```python
# Access the 34th row by position (not by index label)
draft_range.iloc[905]
```

```python
draft_range.info  
```

```python
draft_range.isna().sum() # Check for missing values in the DataFrame
```

```python
print (draft_range['college'].unique())
draft_range['college'].value_counts().tail(50)  # Count occurrences of each unique value in the 'college' column
```

```python
sum_college_counts = (draft_range['college'] == 'North Carolina').sum()
print("Number of players from North Carolina:", sum_college_counts)
```

```python
# Group by 'college' and count the number of players from each college
college_counts = draft_range.groupby('college').size()
college_counts
```

```python
draft_range.tail(10)
```

```python
# Ensure 'draft_number' is numeric for comparison
draft_range_numeric = draft_range.copy()
draft_range_numeric['draft_number'] = pd.to_numeric(draft_range_numeric['draft_number'], errors='coerce')

Age = draft_range_numeric[(draft_range_numeric['age'] <= 20) & (draft_range_numeric['draft_number'] <= 1)]
Age.tail(30)  # Display the filtered rows where age is less than or equal to 19 and draft_number is less than or equal to 1
```

```python
# Convert 'draft_number' to numeric before grouping
draft_range_numeric = draft_range.copy()
draft_range_numeric['draft_number'] = pd.to_numeric(draft_range_numeric['draft_number'], errors='coerce')

df5 = draft_range_numeric.groupby('college').mean(numeric_only=True)  # Group by 'college' and calculate the mean of numeric columns
df5['draft_number'].plot(kind='line')  # Plot the mean draft number for each college
plt.xlabel('College')
plt.ylabel('Mean Draft Number')
plt.title('Mean Draft Number by College')
plt.show()

```
