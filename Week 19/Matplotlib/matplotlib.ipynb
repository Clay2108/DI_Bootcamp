```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()  # Create a figure containing a single axes
ax.plot([1, 2, 3, 4], [1, 4, 2, 3])  # Plot some data on the axes
plt.show()
```

```python
import numpy as np

x = np.linspace(0, 10, 100)
y = x ** 2
plt.plot(x, y)
plt.xlabel('X-axis label')
```

```python


x_points = [1, 1.5, 2, 2.5, 3, 3.5]
y_points = [2, 3, 1, 4, 1, 3]
plt.scatter(x_points, y_points, color='blue', marker='o')   
plt.title('Scatter Plot of Given Points')
plt.xlabel('X-axis')
```

```python
# Exercise 2: Basic Plots
x = np.linspace(0, 10, 20)
y1 = x ** 2
y2 = x ** 1.5
plt.plot(x, y1, label='y = x^2')
plt.plot(x, y2, label='y = x^1.5')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.savefig('basic_plot.png')

```

```python
# Exercise 3: Saving Plots
plt.plot(x, y1, label='y = x^2')
plt.plot(x, y2, label='y = x^1.5')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.savefig('plot.png')  # Save the plot as a PNG file

```

```python
# Exercise 4: Styling Plots
x = np.linspace(0, 10, 20)
y1 = x ** 2
y2 = x ** 1.5
plt.plot(x, y1, label='y = x^2')
plt.plot(x, y2, label='y = x^1.5')
plt.title('My Styled Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

```
