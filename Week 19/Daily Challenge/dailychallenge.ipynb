```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv", encoding='utf-8')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
# check null values
df.isnull().sum()
```

```python
df.dropna(subset=['Age'], inplace=True)
```

```python
# replace null values in Cabin column with 'unknown'
df['Cabin'].fillna('N/A', inplace=True)
df.dropna(subset=['Embarked'], inplace=True)
df.isnull().sum()
```

```python
df['Age']= df['Age'].astype(int)
```

```python
df.head()
```

```python
df.to_csv("titanic_cleaned.csv", index=False)
```

```python
df= pd.read_csv("titanic_cleaned.csv", encoding='utf-8')
df
```

```python
#plot the number of males to females pie chart
data=df['Sex'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(data, labels=data.index, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff'])
plt.legend()
plt.title('Males vs Females on Titanic')
```
