```python
# Exercise 1 : Student Grade Summary
student_grades = {
    "Alice": [88, 92, 100],
    "Bob": [75, 78, 80],
    "Charlie": [92, 90, 85],
    "Dana": [83, 88, 92],
    "Eli": [78, 80, 72]
}
#1
def student_average(grades):
    return sum(grades) / len(grades)

for student, grades in student_grades.items():
    average = student_average(grades)
    print(f"{student}: {average:.2f}")

#2
def student_letter_grade(average):
    if average >= 90:
        return 'A'
    elif 80 <= average < 90:
        return 'B'
    elif 70 <= average < 80:
        return 'C'
    elif 60 <= average < 70:
        return 'D'
    else:
        return 'F'

print("\n")
for student, grades in student_grades.items():
    avg = student_average(grades)
    letter_grade = student_letter_grade(avg)
    print(f"{student}: {letter_grade}")

#3
def class_average(grades_dict):
    total = 0
    count = 0
    for grades in grades_dict.values():
        total += sum(grades)
        count += len(grades)
    return total / count if count > 0 else 0
class_avg = class_average(student_grades)
print(f"\nClass Average: {class_avg:.2f}")    


print("\n")

# #4
max_name_length = max(len(name) for name in student_grades.keys())
for name in student_grades.keys():
    spaces = ' ' * (max_name_length - len(name))
    print(f"{name}:{spaces} Average Grade = {student_averages[name]:.2f}, Letter Grade = {student_letter_grades[name]}")
```

```python
# Exercise 2 : Advanced Data Manipulation and Analysis

sales_data = [
    {"customer_id": 1, "product": "Smartphone", "price": 600, "quantity": 1, "date": "2023-04-03"},
    {"customer_id": 2, "product": "Laptop", "price": 1200, "quantity": 1, "date": "2023-04-04"},
    {"customer_id": 1, "product": "Laptop", "price": 1000, "quantity": 1, "date": "2023-04-05"},
    {"customer_id": 2, "product": "Smartphone", "price": 500, "quantity": 2, "date": "2023-04-06"},
    {"customer_id": 3, "product": "Headphones", "price": 150, "quantity": 4, "date": "2023-04-07"},
    {"customer_id": 3, "product": "Smartphone", "price": 550, "quantity": 1, "date": "2023-04-08"},
    {"customer_id": 1, "product": "Headphones", "price": 100, "quantity": 2, "date": "2023-04-09"},
]

# 1. 
def total_sales_calculation(sales):
    total_sales = {}
    for sale in sales:
        product = sale["product"]
        total = sale["price"] * sale["quantity"]
        if product in total_sales:
            total_sales[product] += total
        else:
            total_sales[product] = total
    return total_sales

print("\n#1 Total Sales for Each Product:")
total_sales = total_sales_calculation(sales_data)
for product, total in total_sales.items():
    print(f"{product}: ${total}")


#2.Customer Spending Profile
def customer_spending_profile(sales):
    spending_profile = {}
    for sale in sales:
        customer_id = sale["customer_id"]
        total = sale["price"] * sale["quantity"]
        if customer_id in spending_profile:
            spending_profile[customer_id] += total
        else:
            spending_profile[customer_id] = total
    return spending_profile
print("\n#2 Customer Spending Profile:")
customer_spending = customer_spending_profile(sales_data)   
for customer_id, total in customer_spending.items():
    print(f"Customer {customer_id}: ${total}")


#3. Sale Data Enhancement:
for transaction in sales_data:
    transaction["total_price"] = transaction["price"] * transaction["quantity"]
print("\n#3 Enhanced Sales Data with Total Price:")
for transaction in sales_data:
    print(transaction)

# 4. High-Value Transactions:
High_value_transactions = [transaction for transaction in sales_data if transaction["total_price"] > 500]
print("\n#4 High-Value Transactions (Total Price > $500):")
# Sort the high value transactions by total_price in descending order
sorted_high_value_transactions = sorted(High_value_transactions, key=lambda x: x["total_price"], reverse=True)
for transaction in sorted_high_value_transactions:
    print(transaction)


#5 Customer Loyalty Identification:
customer_loyalty = {}
for transaction in sales_data:
    customer_id = transaction["customer_id"]
    if customer_id not in customer_loyalty:
        customer_loyalty[customer_id] = 0
    customer_loyalty[customer_id] += 1
print("\n#5 Customer Loyalty Identification (Number of Transactions):")
for customer_id, count in customer_loyalty.items():
    print(f"Customer {customer_id}: {count} transactions") 


```
