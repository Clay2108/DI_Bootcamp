```python
# Exercise 1: Cats

class Cat:
    def __init__(self, cat_name, cat_age):
        self.name = cat_name
        self.age = cat_age

cat1 = Cat("Tom", 3)
cat2 = Cat("Jerry", 2)
cat3 = Cat("Spike", 4)

def find_oldest_cat(cat1, cat2, cat3):
    oldest_cat = cat2
    if cat3.age > oldest_cat.age:
        oldest_cat = cat3
    if cat1.age > oldest_cat.age:
        oldest_cat = cat1
    return oldest_cat   
print(f"The oldest cat is {find_oldest_cat(cat1, cat2, cat3).name} and is {find_oldest_cat(cat1, cat2, cat3).age} years old.")
 
    
```

```python
# Exercise 2 : Dogs
class Dog:
    def __init__(self, dog_name, dog_height):
        self.name = dog_name
        self.height = dog_height
    def bark(self):
        print(f"{self.name} goes woof!")
    def jump(self):
        print(f"{self.name} jumps {self.height * 2} cm high!")

davids_dog = Dog("Miley", 40)
sarahs_dog = Dog("Brownie", 70)
davids_dog.bark()
davids_dog.jump()
sarahs_dog.bark()
sarahs_dog.jump()

compare_dogs= lambda dog1, dog2: dog1 if dog1.height > dog2.height else dog2
print(f"The taller dog is {compare_dogs(davids_dog, sarahs_dog).name} and is {compare_dogs(davids_dog, sarahs_dog).height} cm tall.")
```

```python
# Exercise 3 : Whoâ€™s the song producer?

class Song:
    def __init__(self, song_name, song_lyrics):
        self.name = song_name
        self.lyrics = song_lyrics
    def sing_me_A_song(self):
        print(f"{self.name} lyrics:")
        for line in self.lyrics:
            print(line)

# Example usage:
happy_birthday = Song("Happy Birthday", [
    "Happy birthday to you",
    "Happy birthday to you",
    "Happy birthday dear friend",
    "Happy birthday to you"
])
happy_birthday.sing_me_A_song()
```

```python
# Exercise 4 : Afternoon at the Zoo

class Zoo:
    def __init__(self, zoo_name):
        self.name = zoo_name
        self.animals = []

    # add a method to add an animal
    def add_animal(self, new_animal):
        if new_animal not in self.animals:
            self.animals.append(new_animal)

    def get_animals(self):
        if self.animals:
            print(f"Animals in the zoo:")
            for animal in self.animals:
                print(f"-{animal}")
        else:
            print("No animals in the zoo.")

    def sell_animal(self, animal_sold):
        if animal_sold in self.animals:
            self.animals.remove(animal_sold)
            print(f"{animal_sold} has been sold.")
        else:
            print(f"{animal_sold} is not in the zoo.")

    def sort_animals(self):
        self.animals.sort()
        print("Animals sorted alphabetically:")
        for animal in self.animals:
            print(f"- {animal}")

    def get_groups(self):
        groups = {}
        for animal in self.animals:
            first_letter = animal[0].upper()
            if first_letter not in groups:
                groups[first_letter] = []
            groups[first_letter].append(animal)
        return groups
my_zoo = Zoo("Safari Park")


my_zoo.add_animal("Lion")
my_zoo.add_animal("Tiger")
my_zoo.add_animal("Elephant")
my_zoo.add_animal("Giraffe")
my_zoo.add_animal("Zebra")
my_zoo.add_animal("Eagle")
my_zoo.get_animals()

my_zoo.sell_animal("Tiger")
my_zoo.get_animals()

my_zoo.sort_animals()
my_zoo.get_groups()


# Example usage:
# zoo = Zoo("City Zoo")
# zoo.add_animal("Monkey")
# zoo.add_animal("Lion")
# zoo.add_animal("Baboon")
# zoo.add_animal("Bear")

# zoo.add_animal("Elephant")
# zoo.add_animal("Giraffe")
# zoo.get_animals()

# zoo.add_animal("Zebra")
# zoo.sort_animals()
# zoo.sell_animal("Lion")
# zoo.get_animals()
# zoo.get_groups()


```
